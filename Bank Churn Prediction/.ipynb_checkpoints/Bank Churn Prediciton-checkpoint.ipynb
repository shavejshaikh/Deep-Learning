```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
col_to_scale = ['CreditScore','Age','Tenure','NumOfProducts','EstimatedSalary']

df2[col_to_scale] = scaler.fit_transform(df2[col_to_scale])
```

```python
X = df2.drop('Exited',axis='columns')
Y  =df2.Exited
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
print("{},{}".format(len(x_train),len(x_test)))
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(13,input_shape=(12,),activation='relu'), #Input 
    keras.layers.Dense(13,activation='relu'), #Hidden 
    keras.layers.Dense(1,activation='sigmoid') # Output 
    
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=150)
```

```python
model.evaluate(x_test,y_test)
```
